```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x**2 + 3*y**2 - 3*x - y + 5

def fx(x, y):
    return 2*x - 3

def fy(x, y):
    return 6*y - 1

def df(x,y): 
    return np.array([fx(x,y), fy(x,y)])

x0, y0 = 1.5, 1/6
```

```python
ls = np.linspace(-20,20,400)
X,Y = np.meshgrid(ls,ls)
```

```python
plt.contourf(X, Y, f(X,Y), levels=20)
plt.scatter([x0], [y0], c='g')
plt.gcf().set_size_inches(10,10)
```
